```python
vin_failure_analysis_encoded2
```

```python
plt_heatmap = vin_failure_analysis_encoded2.groupby('Variant').sum().sort_values(by='commandStatus_Failure', ascending=False)
plt_heatmap = plt_heatmap.rename(columns={
    'commandStatus_Failure': 'Failure',
    'commandStatus_Success': 'Success',
    'commandStatus_Timed Out': 'TimedOut',
})  
```

```python
plt_heatmap.loc[' '] = plt_heatmap.sum()
plt_heatmap
```

```python
sns.heatmap(plt_heatmap, cmap='Blues',annot=True,fmt='g',cbar_kws={'label': 'Count'})
```

```python
plt_heatmap_percentage = plt_heatmap.copy()
```

```python
plt_heatmap_percentage['Failure'] = ((plt_heatmap_percentage['Failure'] * 100) / plt_heatmap_percentage['Failure'].sum()).round(2)
plt_heatmap_percentage['Success'] = ((plt_heatmap_percentage['Success'] * 100) / plt_heatmap_percentage['Success'].sum()).round(2)
plt_heatmap_percentage['TimedOut'] = ((plt_heatmap_percentage['TimedOut'] * 100) / plt_heatmap_percentage['TimedOut'].sum()).round(2)
# plt_heatmap_percentage = plt_heatmap_percentage.rename(columns={
#     'commandStatus_Failure': 'Failure',
#     'commandStatus_Success': 'Success',
#     'commandStatus_Timed Out': 'TimedOut',
# })  
```
